```python
from amaranth import *
from amaranth.sim import *
from myhdlpeek.amaranth import Peeker
from amaranth.back import verilog
from steps import Steps

dut = Steps()
with open("steps.v", "w") as f:
    f.write(verilog.convert(dut, ports=dut.ports()))


sim = Simulator(dut)

# Add a 1 MHz clock to the simulator.
sim.add_clock(5e-3)

Peeker.assign_simulator(sim)

# Simulate for 20 microseconds.
sim.run_until(3)

Peeker.show_waveforms()

Peeker.clear()
```
